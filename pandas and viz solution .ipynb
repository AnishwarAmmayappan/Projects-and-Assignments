```python
# 3.Which Team had won by maximum runs?

match.loc[match['win_by_runs']==match['win_by_runs'].max(),['winner','win_by_runs']]
```

```python
match.sort_values(by='win_by_runs',ascending=False).head(10).plot('winner','win_by_runs','bar')
plt.xticks(rotation=90)
```

```python
match.sort_values(by='win_by_runs',ascending=False)
```

```python
# 4.Which Team had won by maximum wicket?

match.loc[match['win_by_wickets']==match['win_by_wickets'].max(),['winner','win_by_wickets']]
```

```python
match.sort_values(by='win_by_wickets',ascending=False).head(20).plot('winner','win_by_wickets','bar')
plt.xticks(rotation=90)
```

```python
# 5.Which Team had won by closest Margin (minimum runs)?
match1=match[match['win_by_runs']>0]

match1.loc[match1['win_by_runs']==match1['win_by_runs'].min(),['winner','win_by_runs']]
```

```python
match1.sort_values(by='win_by_runs',ascending=True).head(10).plot('winner','win_by_runs','bar')
plt.xticks(rotation=90)
```

```python
# 6.Which Team had won by minimum wicket?
match1=match[match['win_by_wickets']>0]
match1.loc[match1['win_by_wickets']==match1['win_by_wickets'].min(),['winner','win_by_wickets']]
```

```python
match1.sort_values(by='win_by_wickets',ascending=True).head(10).plot('winner','win_by_wickets','bar')
plt.xticks(rotation=90)
```

```python
match1.sort_values(by='win_by_wickets',ascending=True)
```

```python
# 9.Has Toss-winning helped in winning matches?

match['toss_based_result']=np.where(match['toss_winner']==match['winner'],'team won match','team lost match')
```

```python
match.head()
```

```python
plt.bar(x=match['toss_based_result'].unique(),height=(match['toss_based_result'].value_counts()/match['id'].count())*100)
plt.xlabel('toss winner')
plt.ylabel('percentage win')
```
